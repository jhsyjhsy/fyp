# Export one sentence per line for CRIE

```python
import pandas as pd
import os

base = "/Users/joannehui/Desktop/fyp/padchest/Reports_public"
out_dir = os.path.join(base, "crie_inputs")
os.makedirs(out_dir, exist_ok=True)

targets = [
    # direct zh
    ("mistral_zh_cn",  "reports_500_mistral_zh.csv",         "mistral_zh_cn"),
    ("mistral_zh_hk",  "reports_500_mistral_zh.csv",         "mistral_zh_hk"),
    ("medgemma_zh_cn", "output_medgemma_zh.csv",             "zh_cn"),
    ("medgemma_zh_hk", "output_medgemma_zh.csv",             "zh_hk"),

    # 2‑step medgemma
    ("medgemma_nllb_Hans", "reports_500_medgemma_nllb_bi.csv", "nllb_zh_Hans_medgemma"),
    ("medgemma_nllb_Hant", "reports_500_medgemma_nllb_bi.csv", "nllb_zh_Hant_medgemma"),
    ("medgemma_opus_v1",   "reports_500_medgemma_opusmt_bi.csv", "opusmt_zh_cn_medgemma_v1"),
    ("medgemma_opus_v2",   "reports_500_medgemma_opusmt_bi.csv", "opusmt_zh_cn_medgemma_v2"),

    # 2‑step mistral
    ("mistral_nllb_cn", "reports_500_translated_mistral_nllb_zh.csv", "lay_zh_cn"),
    ("mistral_nllb_tw", "reports_500_translated_mistral_nllb_zh.csv", "lay_zh_tw"),
    ("mistral_opus_cn", "reports_500_translated_zh_mistral_opus.csv", "opus_lay_zh_cn_like"),
    ("mistral_opus_tw", "reports_500_translated_zh_mistral_opus.csv", "opus_lay_zh_tw_like"),
]

for tag, fname, col in targets:
    path = os.path.join(base, fname)
    df = pd.read_csv(path)
    texts = df[col].fillna("").astype(str)

    # filter out blank lines
    lines = [t.strip().replace("\n", " ") for t in texts if t.strip()]
    out_path = os.path.join(out_dir, f"{tag}.txt")
    with open(out_path, "w", encoding="utf-8") as f:
        for line in lines:
            f.write(line + "\n")

    print(f"Wrote {len(lines)} lines to {out_path}")
```

```python
import pandas as pd
from io import StringIO
import re
import numpy as np

# Paste your full table here
raw_data = """
|Path|檔名|字數|相異詞數比率|難詞數|實詞數|複雜語意類別句子數|實詞密度|領域實詞頻對數平均|斷詞(詞性標記)|備註|
|crie_inputs/|medgemma_nllb_Hans.txt|14200|0.08850681|3120|6585|147|0.7598661|1.307847| | |
|crie_inputs/|medgemma_nllb_Hant.txt|9937|0.1146487|2280|4704|82|0.7563917|1.34761| | |
|crie_inputs/|medgemma_opus_v1.txt|15760|0.0950162|3440|7706|489|0.7805915|1.384515| | |
|crie_inputs/|medgemma_opus_v2.txt|15760|0.0950162|3440|7706|489|0.7805915|1.384515| | |
|crie_inputs/|medgemma_zh_cn.txt|7250|0.1155458|1959|4044|345|0.891731|1.350536| | |
|crie_inputs/|medgemma_zh_hk.txt|8451|0.1062599|2515|5017|526|0.8797124|1.428904| | |
|crie_inputs/|mistral_nllb_cn.txt|51784|0.03909561|11969|25130|685|0.7772726|1.467907| | |
|crie_inputs/|mistral_nllb_tw.txt|43035|0.05057037|9605|21607|571|0.7849669|1.381631| | |
|crie_inputs/|mistral_opus_cn.txt|53332|0.04673468|12173|26933|2079|0.7827768|1.542196| |
"""

# Parse Markdown table to DataFrame
df = pd.read_csv(StringIO(raw_data), sep='|', skipinitialspace=True).dropna(how='all').reset_index(drop=True)
df.columns = df.columns.str.strip()
valid_rows = df[df['字數'] > 0].copy()  # Filter valid data
valid_rows['檔名'] = valid_rows['檔名'].str.strip()
valid_rows['DE'] = pd.to_numeric(valid_rows['相異詞數比率'], errors='coerce')  # Difficulty
valid_rows['VH'] = pd.to_numeric(valid_rows['實詞密度'], errors='coerce')     # Vocab Hardness
valid_rows['Na'] = pd.to_numeric(valid_rows['領域實詞頻對數平均'], errors='coerce')  # Words/sentence
valid_rows['chars'] = pd.to_numeric(valid_rows['字數'], errors='coerce')

print("Raw Data Preview:")
print(valid_rows[['檔名', 'DE', 'VH', 'Na', 'chars']].round(3).head())

# Analysis like English version
print("\nPipeline Averages:")
pipeline_groups = {
    'MedGemma-NLLB': valid_rows[valid_rows['檔名'].str.contains('nllb')],
    'MedGemma-Opus': valid_rows[valid_rows['檔名'].str.contains('opus')],
    'MedGemma-Direct': valid_rows[valid_rows['檔名'].str.contains('medgemma_zh')],
    'Mistral-NLLB': valid_rows[valid_rows['檔名'].str.contains('mistral_nllb')],
    'Mistral-Opus': valid_rows[valid_rows['檔名'].str.contains('mistral_opus')],
    'Mistral-Direct': valid_rows[valid_rows['檔名'].str.contains('mistral_zh')]
}
for group, gdf in pipeline_groups.items():
    if len(gdf) > 0:
        print(f"{group}: DE={gdf['DE'].mean():.3f}, VH={gdf['VH'].mean():.3f}, Na={gdf['Na'].mean():.3f}")

avg_de = [gdf['DE'].mean().round(3) if len(gdf)>0 else np.nan for gdf in pipeline_groups.values()]
avg_vh = [gdf['VH'].mean().round(3) if len(gdf)>0 else np.nan for gdf in pipeline_groups.values()]
avg_na = [gdf['Na'].mean().round(3) if len(gdf)>0 else np.nan for gdf in pipeline_groups.values()]
n_files = [len(gdf) for gdf in pipeline_groups.values()]

comparison = pd.DataFrame({
    'Pipeline': list(pipeline_groups.keys()),
    'Avg DE': avg_de,
    'Avg VH': avg_vh,
    'Avg Na': avg_na,
    'N Files': n_files
})
print("\nSummary Table:\n", comparison.to_string(index=False))

# Easiest pipeline
best_de = valid_rows.loc[valid_rows['DE'].idxmin(), '檔名']
print(f"\nEasiest (lowest DE): {best_de}")
```

# Statistical Tests

```python
import pandas as pd
import numpy as np
from scipy import stats
from io import StringIO

# Use your existing df/valid_rows from before
# Add these stats after the averages table

print("\nStatistical Metrics:")

# 1. Overall descriptives
print("Overall Descriptives:")
desc = valid_rows[['DE', 'VH', 'Na']].describe().round(3)
print(desc)

# 2. Correlations (e.g., DE vs chars)
print("\nCorrelations:")
corr_matrix = valid_rows[['DE', 'VH', 'Na', 'chars']].corr().round(3)
print(corr_matrix)

# 3. ANOVA: Pipeline differences (MedGemma vs Mistral)
pipelines = {
    'MedGemma': valid_rows[valid_rows['檔名'].str.contains('medgemma')],
    'Mistral': valid_rows[valid_rows['檔名'].str.contains('mistral')]
}
de_med = pipelines['MedGemma']['DE']
de_mistral = pipelines['Mistral']['DE']
t_stat, p_val = stats.ttest_ind(de_med, de_mistral)
print(f"\nT-test (MedGemma vs Mistral DE): t={t_stat:.3f}, p={p_val:.3f} (significant if p<0.05)")

# 4. Best/worst per metric
print("\nExtremes:")
print(f"Lowest DE (easiest): {valid_rows.loc[valid_rows['DE'].idxmin(), '檔名']} (DE={valid_rows['DE'].min():.3f})")
print(f"Highest DE (hardest): {valid_rows.loc[valid_rows['DE'].idxmax(), '檔名']} (DE={valid_rows['DE'].max():.3f})")
print(f"Lowest VH (easiest vocab): {valid_rows.loc[valid_rows['VH'].idxmin(), '檔名']} (VH={valid_rows['VH'].min():.3f})")

# 5. Grouped stats with SEM (standard error)
grouped = valid_rows.groupby(valid_rows['檔名'].str.extract(r'(medgemma|mistral)_')[0]).agg({
    'DE': ['mean', 'std', 'sem', 'count'],
    'VH': ['mean', 'std'],
    'Na': ['mean']
}).round(3)
grouped.columns = ['DE_mean', 'DE_std', 'DE_sem', 'DE_n', 'VH_mean', 'VH_std', 'Na_mean']
print("\nGrouped Stats (MedGemma/Mistral):\n", grouped)

# 6. Effect size (Cohen's d for DE)
cohens_d = (de_med.mean() - de_mistral.mean()) / np.sqrt((de_med.std()**2 + de_mistral.std()**2)/2)
print(f"Cohen's d (DE): {cohens_d:.3f} (0.2=small, 0.5=medium, 0.8=large)")

# Export to CSV
valid_rows.to_csv('crie_analysis.csv', index=False)
print("\nExported to crie_analysis.csv")
```
